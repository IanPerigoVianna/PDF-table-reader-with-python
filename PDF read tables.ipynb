```python
#Vamos ler uma tabela da página anterior
tabela_lista2 = tabula.read_pdf('ResultadoVale.pdf', pages= "10")
tabela2 = tabela_lista2[0] 

```

```python
# Vamos atribuir nossas colunas como a linha de índice 0 da tabela
tabela2.columns = tabela2.iloc[0]
```

```python
#vamos criar duas colunas que serão a sepração da coluna "variação percentual"
tabela2[[0,1]]= tabela2["Variação percentual"].str.split(" ", expand= True)
```

```python
display(tabela2)
```

```python
#Vendo o pdf notamos que são 2 tabelas distintas na página 10, porém foi lido pelo tabula como uma tabela só, vamos separar
tabela2 = tabela2[1:9]
tabela2 = tabela2.set_index("R$ milhões")
```

```python
tabela2.columns = tabela2.iloc[0]
tabela2 = tabela2[1:]
```

```python
#Vamos dropar as colunas original sem a separação
tabela2 = tabela2.drop("1T21/4T20 1T21/1T20", axis= 1)
```

```python
display(tabela2)
```

Vamos criar uma função para tratar as vírgulas,parênteses e símbolo de porcentagem.

```python
#Vamos importar o módulo para expressões regulares
import re
def remove_and_convert(value):
    if isinstance(value,str):
        cleaned_value = value.replace(',', '.')
        cleaned_value = cleaned_value.replace('(', '')
        cleaned_value = cleaned_value.replace(')', '')
        cleaned_value = cleaned_value.replace('%', '')
        return float(cleaned_value) / 100 if '%' in value else float(cleaned_value)
    return value
```

```python
tabela2[['1T21','4T20', '1T20']] = tabela2[['1T21','4T20','1T20']].applymap(remove_and_convert)
```

```python
display(tabela2)
```

Agora nossa tabela está pronta para fazermos operações

```python
revenue_20_to_21 = tabela2.iloc[0,0] + tabela2.iloc[0,1] + tabela2.iloc[0,2]
expenses_20_to_21 = tabela2.iloc[1,0] + tabela2.iloc[1,1] + tabela2.iloc[1,2]
```

Somamos as receitas e custos dos períodos

```python
revenue_20_to_21
```

```python
expenses_20_to_21
```

```python
profit = revenue_20_to_21 - expenses_20_to_21
```
